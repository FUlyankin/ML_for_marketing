#  Продуктовые метрики, работа с логами

В этой тетрадке мы проанализируем [логи покупок онлайн-магазина, продающего атрибутику компании Google.](https://www.kaggle.com/c/ga-customer-revenue-prediction/data?select=train.csv)

- Построить несколько пользовательских метрик: посещаемость, средняя выручка, возвращаемость
